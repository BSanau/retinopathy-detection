# Model metrics
Once our model and its corresponding weights are saved, we study its metrics.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from keras.preprocessing.image import ImageDataGenerator
from sklearn.metrics import confusion_matrix
from tensorflow.keras.models import load_model

%matplotlib inline
```

Opening the model's history, saved during the training.

```python
df = pd.read_csv("../models/fun_224_16_softmax_Adam_100_20.csv")
df.drop(columns = "Unnamed: 0", inplace=True)
df.head()
```

### Accuracy
The accuracy tells how often the model is making a correct prediction.

```python
plt.figure(figsize = (10,5))
plt.plot(df["accuracy"])
plt.plot(df["val_accuracy"])
plt.xlabel("epochs")
plt.ylabel("accuracy")
plt.legend(["train", "test"])
plt.title("Model Accuracy");
```

### Loss
The loss represents how bad is the model's prediction. If the model's prediction is perfect, the loss is zero; otherwise, the loss is greater. The goal of training a model is to find a set of weights that have low loss, on average, across all images.

```python
plt.figure(figsize = (10,5))
plt.plot(df["loss"])
plt.plot(df["val_loss"])
plt.xlabel("epochs")
plt.ylabel("loss")
plt.legend(["train", "test"])
plt.title("Model Loss");
```

### Confusion matrix
The confusion matrix allows us to visualize the performance of the algorithm. Each row of the matrix represents the instances in the real class while each column represents the instances in the predicted class. The name stems from the fact that it makes it easy to see if the system is confusing two classes (i.e. commonly mislabeling one as another). 

```python
# Load the model
mymodel = load_model("../models/fun_224_16_softmax_Adam_100_20_structure.h5")
mymodel.load_weights("../models/fun_224_16_softmax_Adam_100_20_weights.hdf5")
```

```python
# Prepare data
train_datagen = ImageDataGenerator(
            # Image standardization
            samplewise_center=True, 
            samplewise_std_normalization=True, 
            )
test_set = train_datagen.flow_from_directory(
    '../../rawdata/test',
    target_size = (224, 224),
    batch_size = 16,
    shuffle = False) # sorts data in alphanumeric order
```

As we are using a functional model, the result will be an array of 4 columns, containing the probability of belonging to each class.

```python
y_pred = mymodel.predict(
    test_set, 
    batch_size = 16, 
    verbose = 1)

# The predicted class will be the one with the highest probability
y_classes = y_pred.argmax(axis=1)
```

Calculating confusion matrix:

```python
cm = confusion_matrix(test_set.classes, y_classes, normalize = "true")
```

```python
sns.heatmap(cm, cmap='viridis', annot = True)
plt.xticks(range(0,4), ["CNV", "DME", "DRUSEN", "NORMAL"])
plt.xlabel("predicted")
plt.yticks(range(0,4), ["CNV", "DME", "DRUSEN", "NORMAL"])
plt.ylabel("real")
plt.title ("Confusion matrix");
```
